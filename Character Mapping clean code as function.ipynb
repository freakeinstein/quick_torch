### Character Mapping clean code as function

global variable to hold every datastructure that emma have

```lua
emma = {}
emma.char_map = {}
modules = {}
```

```lua
function emma.char_map:initialize_mapping(lang)
        
    emma.char_map.ASCII_UNICODE = {}
    emma.char_map.UNICODE_ASCII = {}
    
    if lang == 'mal' then
        print("setting up malayalam mapping")
    
        for i=0,127 do
        emma.char_map.ASCII_UNICODE[string.char(i)] = string.char(i)
        emma.char_map.UNICODE_ASCII[string.char(i)] = string.char(i)
        end
        local x,y,z = 224,180,128
        for i=128,255 do
        emma.char_map.ASCII_UNICODE[string.char(i)] = string.char(x,y,z)
        emma.char_map.UNICODE_ASCII[string.char(x,y,z)] = string.char(i)
        z = z+1 ;
        if z > 191 then z,y = 128,y+1 end
        end 
    end
    
    function emma.char_map.A2U(character)
        return emma.char_map.ASCII_UNICODE[character]
    end
    
    function emma.char_map.U2A(character)
        return emma.char_map.UNICODE_ASCII[character]
    end
end
```

## Documentation:       
### initialize_mapping([string language])

- specify the language ( ex: malayalam : mal ) as argument.
> By default, char_map will be initialized with two mapping tables,  ASCII_UNICODE and UNICODE_ASCII and eack of them will contain corresponding data related to the language we set. I also will be initialized with functions, A2U(character_for_conversion) (ASCII to UNICODE) and U2A(character_for_conversion) (UNICODE TO ASCII) these will return the corresponding values from ASCII_UNICODE and UNICODE_ASCII tables.

```lua
function modules:mapping_module(lan)
    emma.char_map:initialize_mapping(lan)
    return emma.char_map
end
```

## Documentation: 
### [emma.char_map] modules.mapping_module([string language]) 
- specify the language ( ex: malayalam : mal ) as argument.
- will setup the first module for mapping functionalities
- returns the **emma.char_map** object, which deals with all basic character mapping functions and datastructures.

```lua
map = modules:mapping_module('mal')
```

```lua
print(string.byte(map.U2A('ം')),map.A2U(string.char(130)))
```

```lua
print(string.byte(map.U2A('ം')),map.A2U(string.char(100)))
```

```lua
file_in = io.open('data/malayalam_in.txt','r')
file_out = io.open('data/malayalam_out_encoded.txt','a')

local map = modules:mapping_module('mal')

for i in file_in:lines() do -- get each line from the file --
    for code in i:gmatch("[%z\1-\127\194-\244][\128-\191]*") do -- get each character from the line above --
        encode = map.U2A(code)
        if encode then file_out:write(encode)
        else print(encode,string.byte(code)) end
    end 
end

file_in:close()
file_out:close()
```
